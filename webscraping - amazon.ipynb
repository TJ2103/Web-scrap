```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.amazon.com/dp/B0C1J4MPBB/'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
# Find the element containing the number of ratings
soup = BeautifulSoup(response.text, 'html.parser')
ratings_element = soup.find('span', {'id': 'acrCustomerReviewText'})

# Find the elements containing the data you want to scrape
title_element = soup.find('span', {'id': 'productTitle'})
price_element = soup.find('span', {'class': 'a-price-whole'})
price_cent_element = soup.find('span', {'class': 'a-price-fraction'}) #.find('span', {'id': 'acrPopover'})
rating_element = soup.find('div', {'id': 'centerCol'}).find('div',{'id':'averageCustomerReviews'}).find('a', {'class':'a-popover-trigger'}).find('span',{'class':'a-color-base'})
rating_count_element = soup.find('div', {'id': 'averageCustomerReviews_feature_div'}).find('span', {'id': 'acrCustomerReviewText'})
image_element = soup.find('img', {'id': 'landingImage'})
brand_element = soup.find('tr', {'class': 'po-brand'}).find('span', {'class': 'po-break-word'})
screensize_element = soup.find('tr', {'class': 'po-display.size'}).find('span', {'class': 'po-break-word'})
internet_service_supported_element = soup.find('tr', {'class': 'po-supported_internet_services'}).find('span', {'class': 'po-break-word'})
display_technology_element = soup.find('tr', {'class': 'po-display.technology'}).find('span', {'class': 'po-break-word'})
product_dimensions_element = soup.find('tr', {'class': 'po-item_depth_width_height'}).find('span', {'class': 'po-break-word'})
resolution_element = soup.find('tr', {'class': 'po-resolution'}).find('span', {'class': 'po-break-word'})
refresh_rate_element = soup.find('tr', {'class': 'po-refresh_rate'}).find('span', {'class': 'po-break-word'})
special_feature_element = soup.find('tr', {'class': 'po-special_feature'}).find('span', {'class': 'po-break-word'})
model_name_element = soup.find('tr', {'class': 'po-model_name'}).find('span', {'class': 'po-break-word'})
included_components_element = soup.find('tr', {'class': 'po-included_components'}).find('span', {'class': 'po-break-word'})

# Extract the text from the elements
title = title_element.text.strip() if title_element else '-'
price = price_element.text.strip() if price_element else '-'
price_cents = price_cent_element.text.strip() if price_cent_element else '-'
rating = rating_element.text.strip() if rating_element else '-'
rating_count = rating_count_element.text.strip() if rating_count_element else '-'
image = image_element.get('src',' ') if image_element else '-'
brand = brand_element.text.strip() if brand_element else '-'
screensize = screensize_element.text.strip() if screensize_element else '-'
internet_service_supported = internet_service_supported_element.text.strip() if internet_service_supported_element else '-'
display_technology = display_technology_element.text.strip() if display_technology_element else '-'
product_dimensions = product_dimensions_element.text.strip() if product_dimensions_element else '-'
resolution = resolution_element.text.strip() if resolution_element else '-'
refresh_rate = refresh_rate_element.text.strip() if refresh_rate_element else '-'
special_feature = special_feature_element.text.strip() if special_feature_element else '-'
model_name = model_name_element.text.strip() if model_name_element else '-'
included_components = included_components_element.text.strip() if included_components_element else '-'

output_json = {
    "name": title,
    "rating": rating,
    "no-of-rating": rating_count.replace('ratings','').strip(),
    "image": image,
    "price": price + price_cents,
    "additional_information":{
        "brand-name": brand,
        "screen-size": screensize,
        "supported-internet-services": internet_service_supported,
        "display-technology": display_technology,
        "product-dimensions": product_dimensions,
        "resolution": resolution,
        "refresh-rate": refresh_rate,
        "special-feature": special_feature,
        "model-name": model_name,
        "included-components": included_components,
    }
}
```
